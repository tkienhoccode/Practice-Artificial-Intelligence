#III. Bài tập ở lớp
Bài 2. Cài đặt thuật giải A* tìm đường đi giữa 2 đỉnh bất kỳ trong đồ thị.

```
import heapq

def a_star_graph(graph, h, start, goal):
    pq = [(h[start], 0, start, [start])]
    visited = set()

    while pq:
        f, g, current, path = heapq.heappop(pq)

        if current == goal:
            print(f"Đường đi từ {path[0]} đến {goal}: {path} (Chi phí: {g})")
            return

        if current in visited:
            continue
        visited.add(current)

        for neighbor, weight in graph[current]:
            if neighbor not in visited:
                new_g = g + weight
                new_f = new_g + h[neighbor]
                heapq.heappush(pq, (new_f, new_g, neighbor, path + [neighbor]))

    print("Không tìm thấy đường đi")

graph_data = {
    'A': [('B', 1), ('C', 3)],
    'B': [('D', 3), ('E', 1)],
    'C': [('F', 5)],
    'D': [],
    'E': [('F', 1)],
    'F': []
}
h_data = {'A': 5, 'B': 3, 'C': 4, 'D': 3, 'E': 1, 'F': 0}

a_star_graph(graph_data, h_data, 'A', 'F')
```

#IV. Bài tập về nhà
Bài 2. Cài đặt bài toán người giao hàng theo thuật giải A*.

```
import heapq

def a_star_tsp(graph, start):
    initial_visited = tuple(sorted([start]))
    pq = [(0, 0, start, initial_visited, [start])]

    n_nodes = len(graph)
    min_cost = float('inf')
    best_path = None

    while pq:
        f, g, u, visited, path = heapq.heappop(pq)

        if len(visited) == n_nodes:
            if start in graph[u]:
                total_cost = g + graph[u][start]
                if total_cost < min_cost:
                    min_cost = total_cost
                    best_path = path + [start]
            continue

        for v, weight in graph[u].items():
            if v not in visited:
                new_visited = tuple(sorted(list(visited) + [v]))
                new_g = g + weight
                new_f = new_g + 0 # h = 0
                heapq.heappush(pq, (new_f, new_g, v, new_visited, path + [v]))

    if best_path:
        print(f"Lộ trình người giao hàng: {' -> '.join(best_path)}")
        print(f"Tổng chi phí nhỏ nhất: {min_cost}")
    else:
        print("Không tồn tại đường đi chu trình.")

tsp_graph = {
    'A': {'B': 10, 'C': 15, 'D': 20},
    'B': {'A': 10, 'C': 35, 'D': 25},
    'C': {'A': 15, 'B': 35, 'D': 30},
    'D': {'A': 20, 'B': 25, 'C': 30}
}

a_star_tsp(tsp_graph, 'A')
```
