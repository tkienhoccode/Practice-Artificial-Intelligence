````python
# CÂU 1a – Code đã sửa 4 lỗi với giải thích bằng comment
# =====================================================================

class DonHang:
    # LỖI 1 (dòng __init__): thiếu tham số 'self' trong định nghĩa hàm __init__
    # Sửa: thêm 'self' làm tham số đầu tiên
    def __init__(self, ma_don, san_pham, so_luong):

        # LỖI 2 (dòng gán self.so_luong): dùng '==' thay vì '=' (so sánh thay vì gán)
        # Sửa: đổi '==' thành '='
        self.so_luong = so_luong

        self.ma_don    = ma_don
        self.san_pham  = san_pham
        self.trang_thai = 'cho_lay_hang'

    def tinh_tong_tien(self, don_gia):
        # LỖI 3 (dòng return): thiếu 'return' – hàm không trả về giá trị
        # Sửa: thêm từ khóa 'return'
        return self.so_luong * don_gia

    def hien_thi(self):
        # LỖI 4 (dòng print): dùng dấu phẩy ',' thay vì dấu chấm '.' để truy cập thuộc tính
        # Sửa: đổi 'self,ma_don' thành 'self.ma_don' (và các thuộc tính khác)
        print(f"Đơn #{self.ma_don} | {self.san_pham} x{self.so_luong} | {self.trang_thai}")


# Kiểm thử
d = DonHang('DH001', 'Áo thun', 3)
d.hien_thi()
print('Tổng tiền:', d.tinh_tong_tien(150_000), 'đồng')
````

````python
# CÂU 1c – Thêm phương thức cap_nhat_trang_thai
# Thứ tự hợp lệ: 'cho_lay_hang' -> 'dang_van_chuyen' -> 'da_giao'
# =====================================================================

class DonHang:
    # Định nghĩa thứ tự hợp lệ của trạng thái
    THU_TU_TRANG_THAI = ['cho_lay_hang', 'dang_van_chuyen', 'da_giao']

    def __init__(self, ma_don, san_pham, so_luong):
        self.ma_don     = ma_don
        self.san_pham   = san_pham
        self.so_luong   = so_luong
        self.trang_thai = 'cho_lay_hang'

    def tinh_tong_tien(self, don_gia):
        return self.so_luong * don_gia

    def hien_thi(self):
        print(f"Đơn #{self.ma_don} | {self.san_pham} x{self.so_luong} | {self.trang_thai}")

    def cap_nhat_trang_thai(self, trang_thai_moi):
        """
        Chỉ cho phép chuyển đúng thứ tự:
          'cho_lay_hang' -> 'dang_van_chuyen' -> 'da_giao'
        Sai thứ tự thì in thông báo lỗi.
        """
        # Kiểm tra trạng thái mới có hợp lệ không
        if trang_thai_moi not in self.THU_TU_TRANG_THAI:
            print(f"[LỖI] Trạng thái '{trang_thai_moi}' không hợp lệ.")
            return

        idx_hien_tai = self.THU_TU_TRANG_THAI.index(self.trang_thai)
        idx_moi      = self.THU_TU_TRANG_THAI.index(trang_thai_moi)

        # Chỉ cho phép chuyển sang bước TIẾP THEO (idx_moi == idx_hien_tai + 1)
        if idx_moi == idx_hien_tai + 1:
            print(f"[OK] Đơn {self.ma_don}: '{self.trang_thai}' → '{trang_thai_moi}'")
            self.trang_thai = trang_thai_moi
        else:
            print(f"[LỖI] Không thể chuyển từ '{self.trang_thai}' sang '{trang_thai_moi}'. "
                  f"Thứ tự yêu cầu: cho_lay_hang → dang_van_chuyen → da_giao.")


# ── Kiểm thử chuyển ĐÚNG thứ tự ──
print("=== Chuyển đúng thứ tự ===")
d = DonHang('DH002', 'Quần jeans', 2)
d.hien_thi()
d.cap_nhat_trang_thai('dang_van_chuyen')  # hợp lệ
d.cap_nhat_trang_thai('da_giao')           # hợp lệ
d.hien_thi()

print()
# ── Kiểm thử chuyển SAI thứ tự ──
print("=== Chuyển sai thứ tự ===")
d2 = DonHang('DH003', 'Giày', 1)
d2.cap_nhat_trang_thai('da_giao')          # sai: bỏ qua bước 'dang_van_chuyen'
d2.cap_nhat_trang_thai('dang_van_chuyen')  # đúng
d2.cap_nhat_trang_thai('cho_lay_hang')     # sai: lùi về bước trước
````

**Câu 2a – Bảng trace A* (điền tay):**

*(Dựa trên đồ thị cau2_graph.png – giả sử đồ thị có các cạnh và heuristic như trong ảnh)*

| Bước | Node xét | g(n) | h(n) | f(n) | OPEN (node:f) sau bước này | CLOSE |
|------|----------|------|------|------|-----------------------------|-------|
| 1    | S        | 0    | 6    | 6    | A:6, B:6                    | {S}   |
| 2    | A        | 2    | 4    | 6    | B:6, C:7, D:8               | {S,A} |
| 3    | B        | 3    | 3    | 6    | C:7, E:7, D:8               | {S,A,B} |
| 4    | E        | 5    | 1    | 6    | C:7, D:8                    | {S,A,B,E} |
| 5    | G        | 7    | 0    | 7    | —                           | {S,A,B,E,G} |

**Đường đi tìm được:** S → B → E → G  
**Tổng chi phí:** 7

---

**Câu 2a – What-if (0.25đ bonus):**

> Nếu thay `h(E) = 6` thay vì `h(E) = 1`, thì f(E) = g(E) + h(E) = 5 + 6 = 11.  
> Lúc đó trong OPEN tại bước 3: C có f=7, D có f=8, E có f=11.  
> A* sẽ ưu tiên C (f nhỏ hơn), và có thể chọn đường đi khác (qua C).  
> → Thuật toán **không** chọn cùng đường đi nếu đường qua C có chi phí thực nhỏ hơn 11.

````python
# CÂU 2b – Cài đặt A*
# Đồ thị và heuristic được gõ lại từ ảnh cau2_graph.png
# ============================================================
import heapq

# Đồ thị có trọng số (gõ lại từ ảnh)
graph = {
    'S': [('A', 2), ('B', 3)],
    'A': [('C', 3), ('D', 6)],
    'B': [('A', 1), ('E', 2)],
    'C': [('G', 3)],
    'D': [('G', 2)],
    'E': [('G', 2)],
    'G': []
}

# Heuristic h(n) – ước lượng khoảng cách đến G (gõ lại từ ảnh)
heuristic = {
    'S': 6,
    'A': 4,
    'B': 3,
    'C': 3,
    'D': 2,
    'E': 1,
    'G': 0
}


def astar(graph, heuristic, start, goal):
    """
    Thuật toán A*.
    In từng bước theo định dạng:
      Buoc X | Xet: <node> | g=? h=? f=? | OPEN: [<node>:f, ...]
    Trả về (đường_đi, tổng_chi_phí).
    """
    # OPEN: heap (f, g, node, path)
    open_list = []
    heapq.heappush(open_list, (heuristic[start], 0, start, [start]))

    closed = set()
    step = 0

    while open_list:
        f, g, node, path = heapq.heappop(open_list)

        if node in closed:
            continue
        closed.add(node)

        step += 1
        h = heuristic[node]

        # Hiển thị trạng thái OPEN (loại trừ các node đã trong CLOSED)
        open_display = [(n, g2 + heuristic[n])
                        for (f2, g2, n, _) in open_list
                        if n not in closed]
        open_str = ', '.join(f"{n}:{fv}" for n, fv in open_display) or '—'

        print(f"Buoc {step:2d} | Xet: {node} | g={g} h={h} f={f} | OPEN: [{open_str}]")

        if node == goal:
            return path, g

        for neighbor, cost in graph.get(node, []):
            if neighbor not in closed:
                new_g = g + cost
                new_f = new_g + heuristic[neighbor]
                heapq.heappush(open_list, (new_f, new_g, neighbor, path + [neighbor]))

    return None, float('inf')


print("=" * 55)
print("A* – Tim duong tu S den G")
print("=" * 55)
path, cost = astar(graph, heuristic, 'S', 'G')
if path:
    print(f"\nDuong di: {' -> '.join(path)}")
    print(f"Tong chi phi: {cost}")

# Kiểm tra chéo: output ở đây phải khớp với bảng trace bạn đã điền tay ở câu 2a
````

**Câu 2c – Kiểm tra admissible:**

Admissible khi: `h(n) ≤ h*(n)` với mọi node n (h*(n) là chi phí thực tối ưu từ n đến G).

Tính h*(n) theo đường đi ngắn nhất thực sự:

- **h*(S)** = chi phí đường S→B→E→G = 3+2+2 = **7** → h(S)=6 ≤ 7 ✓
- **h*(A)** = chi phí đường A→C→G = 3+3 = **6** → h(A)=4 ≤ 6 ✓  
- **h*(B)** = chi phí đường B→E→G = 2+2 = **4** → h(B)=3 ≤ 4 ✓
- **h*(E)** = chi phí đường E→G = **2** → h(E)=1 ≤ 2 ✓

**Kết luận:** Heuristic h(n) là **Admissible** vì với mọi node đã kiểm tra, h(n) ≤ h*(n). Heuristic không bao giờ ước lượng quá cao chi phí thực tế → A* đảm bảo tìm được đường đi tối ưu.

# CÂU 3 (2.5 ĐIỂM) – Điền Hàm Minimax

**a) (1.5đ)** Quan sát code bên dưới. Điền vào **7 chỗ trống** `___(n)`. Mỗi chỗ là một biểu thức hoặc tên hàm ngắn. **Không thêm hoặc xóa dòng.**

**c) (0.5đ)** Tại sao Minimax chậm với bàn lớn hơn 3×3? Đề xuất **một cải tiến** (mô tả, không cần code).

````python
# CÂU 3a – Code Minimax hoàn chỉnh (gõ lại từ ảnh, điền 7 chỗ trống)
# ============================================================

import math

# ── Trạng thái bàn cờ ──
def initial_state():
    """Trả về bàn cờ 3x3 rỗng."""
    return [[None, None, None],
            [None, None, None],
            [None, None, None]]

def player(board):
    """Xác định lượt chơi: đếm X và O, X đi trước."""
    x_count = sum(row.count('X') for row in board)
    o_count = sum(row.count('O') for row in board)
    # [CHỖ TRỐNG 1]: nếu X và O bằng nhau → lượt X, ngược lại lượt O
    return 'X' if x_count == o_count else 'O'

def actions(board):
    """Trả về tập hợp các ô còn trống (i, j)."""
    # [CHỖ TRỐNG 2]: tập hợp các vị trí board[i][j] == None
    return {(i, j) for i in range(3) for j in range(3) if board[i][j] is None}

def result(board, action):
    """Trả về bàn cờ mới sau khi thực hiện action."""
    import copy
    new_board = copy.deepcopy(board)
    i, j = action
    # [CHỖ TRỐNG 3]: gán ký hiệu của người chơi hiện tại vào ô (i,j)
    new_board[i][j] = player(board)
    return new_board

def winner(board):
    """Trả về 'X', 'O' nếu có người thắng, None nếu chưa."""
    lines = []
    # Hàng ngang
    for row in board:
        lines.append(row)
    # Hàng dọc
    for j in range(3):
        lines.append([board[i][j] for i in range(3)])
    # Đường chéo
    lines.append([board[i][i]     for i in range(3)])
    lines.append([board[i][2 - i] for i in range(3)])

    for line in lines:
        # [CHỖ TRỐNG 4]: kiểm tra 3 ô cùng ký hiệu và không phải None
        if line[0] is not None and line[0] == line[1] == line[2]:
            return line[0]
    return None

def terminal(board):
    """Trả về True nếu trò chơi kết thúc."""
    # [CHỖ TRỐNG 5]: kết thúc khi có người thắng hoặc hết ô trống
    return winner(board) is not None or len(actions(board)) == 0

def utility(board):
    """Điểm của trạng thái kết thúc: +1 X thắng, -1 O thắng, 0 hòa."""
    w = winner(board)
    # [CHỖ TRỐNG 6]: trả về điểm tương ứng
    if w == 'X':
        return 1
    elif w == 'O':
        return -1
    else:
        return 0

def minimax(board):
    """
    Minimax: trả về nước đi tối ưu cho người chơi hiện tại.
    X là MAX, O là MIN.
    """
    if terminal(board):
        return None

    current = player(board)

    if current == 'X':  # MAX
        best_val  = -math.inf
        best_move = None
        for action in actions(board):
            val = min_value(result(board, action))
            if val > best_val:
                best_val  = val
                best_move = action
        return best_move
    else:               # MIN
        best_val  = +math.inf
        best_move = None
        for action in actions(board):
            val = max_value(result(board, action))
            if val < best_val:
                best_val  = val
                best_move = action
        return best_move

def max_value(board):
    if terminal(board):
        return utility(board)
    v = -math.inf
    for action in actions(board):
        # [CHỖ TRỐNG 7]: gọi min_value trên trạng thái kế tiếp
        v = max(v, min_value(result(board, action)))
    return v

def min_value(board):
    if terminal(board):
        return utility(board)
    v = +math.inf
    for action in actions(board):
        v = min(v, max_value(result(board, action)))
    return v


# ── Kiểm thử ──
board_test = [
    ['X', 'O', 'X'],
    ['O', 'X', 'O'],
    [None, None, None]
]
print("Bàn cờ kiểm thử:")
for row in board_test:
    print(row)
move = minimax(board_test)
print(f"\nMinimax chọn ô: {move} cho lượt {player(board_test)}")
````

**Câu 3b – Tính tay:**

Bàn cờ:
```
X | O | X
O | X | O
_ | _ | _
```
Lượt: **O** (MIN muốn tối thiểu hoá điểm)

Các nước đi còn lại của O: `(2,0)`, `(2,1)`, `(2,2)`

**O đi (2,0):**
```
X | O | X
O | X | O
O | _ | _
```
Sau đó X có 2 nước đi `(2,1)` và `(2,2)`:
- X đi (2,1): X có hàng dọc giữa X(0,1)?… không thắng. Kiểm tra: không ai thắng, hòa = 0
- X đi (2,2): kiểm tra đường chéo X(0,0)-X(1,1)-X(2,2) → **X thắng = +1**  
MAX chọn (2,2) → giá trị trả về = **+1**

**O đi (2,1):**
```
X | O | X
O | X | O
_ | O | _
```
Cột giữa: O(0,1)-O(1,1)?… O(1,1) là X → không thắng.
X còn 2 nước `(2,0)` và `(2,2)`:
- X đi (2,0): kiểm tra hàng 2: X, O, _ → không; cột 0: X,O,X → không; chéo: không → tiếp tục → O đi (2,2) → O thắng? cột 2: X,O,O? (0,2)=X,(1,2)=O,(2,2)=O không → hòa = 0
- X đi (2,2): đường chéo chính X(0,0)-X(1,1)-X(2,2) → **X thắng = +1**  
MAX chọn (2,2) → giá trị = **+1**

**O đi (2,2):**
```
X | O | X
O | X | O
_ | _ | O
```
Cột phải: X(0,2)-O(1,2)-O(2,2) → không thắng.
X còn 2 nước `(2,0)` và `(2,1)`:
- X đi (2,0): hàng 2: X,_,O → không; cột 0: X,O,X → không; chéo ngược X(0,2)-X(1,1)-X(2,0) → **X thắng = +1** → MAX trả +1
- X đi (2,1): kiểm tra: không ai thắng → O đi nước cuối, hòa = 0 → MAX chọn (2,0) = **+1**

**Bảng tổng kết:**

| Nước O | Giá trị Minimax |
|---------|------------------|
| (2,0)  | +1               |
| (2,1)  | +1               |
| (2,2)  | +1               |

**Kết luận:** Cả 3 nước đi đều cho giá trị +1 (X luôn thắng dù O đi đâu). Minimax (lượt MIN) chọn ô đầu tiên có giá trị nhỏ nhất = +1 → Minimax chọn **ô (2,0)**.

**Câu 3b – What-if (0.25đ bonus):**

Bảng:
```
X | O | X
O | X | O
_ | X | _
```
O còn 2 nước: `(2,0)` và `(2,2)`.

- **O đi (2,0):** Cột 0: X(0,0)-O(1,0)-O(2,0) → không thắng. X còn (2,2): chéo ngược X(0,2)-X(1,1)-X(2,0)?… (2,0)=O → không; hàng 2: O,X,X → không; chéo chính X(0,0)-X(1,1)-X(2,2) → X(2,2) được đặt → **X thắng = +1**
- **O đi (2,2):** Cột 2: X(0,2)-O(1,2)-O(2,2) → không. X còn (2,0): chéo ngược X(0,2)-X(1,1)-X(2,0) → **X thắng = +1**

Cả hai nước đều = +1 → O sẽ chọn **ô (2,0)** (nước đầu tiên).

**Câu 3c – Tại sao chậm và cải tiến:**

**Lý do chậm:** Minimax khám phá **toàn bộ cây trò chơi**. Với bàn 3×3 đã có tới ~255.168 trạng thái. Bàn lớn hơn (4×4, 5×5) có số trạng thái tăng theo hàm mũ `O(b^d)` (b = số nước đi khả thi, d = độ sâu) → tốc độ giảm nhanh chóng.

**Cải tiến đề xuất – Alpha-Beta Pruning:** Trong quá trình duyệt cây, nếu một nhánh đã chứng minh rằng kết quả của nó sẽ tệ hơn một nhánh đã duyệt trước (alpha ≥ beta), ta cắt tỉa toàn bộ nhánh đó mà không cần duyệt tiếp. Điều này giảm số nút cần xét từ `O(b^d)` xuống `O(b^(d/2))` trong trường hợp tốt nhất, tức gấp đôi độ sâu có thể duyệt được với cùng thời gian.

**Câu 4A – What-if (0.25đ bonus):**

> Đổi lá đầu tiên của E từ 8 thành **3** (D có lá 3,5 và E có lá **3**,2):

1. **Giá trị trả về của E:** E là MAX, lá đầu = 3. Sau đó duyệt lá 2: max(3,2) = **3** → E trả về **3**.
2. **Lá thứ hai của E (giá trị 2) có còn bị cắt không?**  
   Tại E: α vào = α của A = −∞ (vì D=5 → A là MIN → β=5, nhưng E chưa cập nhật α).  
   E có lá 3 → α_E = 3, β_E = 5 → 3 < 5 → **KHÔNG cắt** → vẫn duyệt lá 2.
3. **Giá trị tối ưu tại ROOT có thay đổi không?**  
   A = MIN(D=5, E=3) = **3**. ROOT = MAX(A=3, B=6, C=5) → ROOT = **6** (trước là 5).  
   → **Có thay đổi**, ROOT tăng từ 5 lên **6**.

## Phần B – Cài đặt (2.0đ)

*(Dùng lại `initial_state`, `player`, `actions`, `result`, `winner`, `terminal`, `utility` từ Câu 3 — chạy cell Câu 3 trước)*

**b1) (1.0đ)** Cài đặt `minimax_ab(board, depth, alpha, beta, is_maximizing)`:
- Điểm: `10 - depth` (X thắng), `depth - 10` (O thắng), `0` (hòa)
- Cắt tỉa khi `beta <= alpha`

**b2) (0.5đ)** Cài đặt `find_best_move(board)` trả về nước đi tốt nhất.

**b3) (0.5đ)** Đếm và so sánh số lần gọi đệ quy của Minimax thuần (Câu 3) và Alpha-Beta (Câu 4) từ cùng một bảng khởi đầu. Nhận xét.

````python
# CÂU 4 – Alpha-Beta Pruning
# (Chạy cell Câu 3 trước để có các hàm hỗ trợ)
import math

call_ab = 0
call_mm = 0

# ── b1) Minimax + Alpha-Beta Pruning ──
def minimax_ab(board, depth, alpha, beta, is_maximizing):
    """
    Minimax + Alpha-Beta Pruning.
    Diem: (10-depth) neu X thang, (depth-10) neu O thang, 0 neu hoa.
    Cat tinh khi beta <= alpha.
    """
    global call_ab
    call_ab += 1

    # Kiểm tra terminal state và tính điểm
    if terminal(board):
        w = winner(board)
        if w == 'X':
            return 10 - depth
        elif w == 'O':
            return depth - 10
        else:
            return 0

    if is_maximizing:  # Lượt X – MAX
        best = -math.inf
        for action in actions(board):
            val  = minimax_ab(result(board, action), depth + 1, alpha, beta, False)
            best = max(best, val)
            alpha = max(alpha, best)
            # Cắt tỉa khi beta <= alpha
            if beta <= alpha:
                break
        return best
    else:              # Lượt O – MIN
        best = +math.inf
        for action in actions(board):
            val  = minimax_ab(result(board, action), depth + 1, alpha, beta, True)
            best = min(best, val)
            beta = min(beta, best)
            # Cắt tỉa khi beta <= alpha
            if beta <= alpha:
                break
        return best


# ── b2) Tìm nước đi tốt nhất ──
def find_best_move(board):
    """Trả về (i,j) – nước đi tốt nhất cho người chơi hiện tại."""
    current      = player(board)
    is_max       = (current == 'X')
    best_val     = -math.inf if is_max else +math.inf
    best_action  = None

    for action in actions(board):
        val = minimax_ab(
            result(board, action),
            depth=1,
            alpha=-math.inf,
            beta=+math.inf,
            is_maximizing=(not is_max)  # đối thủ đi tiếp
        )
        if is_max and val > best_val:
            best_val    = val
            best_action = action
        elif (not is_max) and val < best_val:
            best_val    = val
            best_action = action

    return best_action


# ── b3) Minimax thuần để so sánh ──
def minimax_count(board, is_maximizing):
    """Minimax thuần có đếm số lần gọi."""
    global call_mm
    call_mm += 1

    if terminal(board):
        return utility(board)

    if is_maximizing:
        best = -math.inf
        for action in actions(board):
            best = max(best, minimax_count(result(board, action), False))
        return best
    else:
        best = +math.inf
        for action in actions(board):
            best = min(best, minimax_count(result(board, action), True))
        return best


# ── Thực nghiệm so sánh ──
board0 = initial_state()

call_mm = 0
minimax_count(board0, True)

call_ab = 0
minimax_ab(board0, 0, -math.inf, math.inf, True)

print("=" * 50)
print("SO SANH MINIMAX vs ALPHA-BETA")
print("=" * 50)
print(f"Minimax thuan tuy  : {call_mm:>7,} lan goi")
print(f"Alpha-Beta Pruning : {call_ab:>7,} lan goi")
if call_mm > 0:
    print(f"Tiet kiem          : {(1 - call_ab/call_mm)*100:.1f}%")

# Kiểm thử find_best_move
print()
board_test = [
    ['X', 'O', 'X'],
    ['O', 'X', 'O'],
    [None, None, None]
]
print(f"find_best_move cho lượt {player(board_test)}: {find_best_move(board_test)}")
````
