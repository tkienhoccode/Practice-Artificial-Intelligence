# II. Bài tập hướng dẫn mẫu
a.	Thuật giải Minimax ứng dụng vào bài toán TicTacToe

```
import copy
import math
import random
import numpy

X = "X"
O = "O"
EMPTY = None
user = None
ai = None

def initial_state():
    """
    Returns starting state of the board.
    """
    return [[EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY]]

def player(board):
    """
    Returns player who has the next turn on a board.
    """
    count = 0
    for i in board:
        for j in i:
            if j:
                count += 1
    if count % 2 != 0:
        return ai
    return user

def actions(board):
    """
    Returns set of all possible actions (i, j) available on the board.
    """
    res = set()
    board_len = len(board)
    for i in range(board_len):
        for j in range(board_len):
            if board[i][j] == EMPTY:
                res.add((i, j))
    return res

def result(board, action):
    """
    Returns the board that results from making move (i, j) on the board.
    """
    curr_player = player(board)
    result_board = copy.deepcopy(board)
    (i, j) = action
    result_board[i][j] = curr_player
    return result_board

def get_horizontal_winner(board):
    # check horizontally
    winner_val = None
    board_len = len(board)
    for i in range(board_len):
        winner_val = board[i][0]
        for j in range(board_len):
            if board[i][j] != winner_val:
                winner_val = None
        if winner_val:
            return winner_val
    return winner_val

def get_vertical_winner(board):
    # check vertically
    winner_val = None
    board_len = len(board)
    for i in range(board_len):
        winner_val = board[0][i]
        for j in range(board_len):
            if board[j][i] != winner_val:
                winner_val = None
        if winner_val:
            return winner_val
    return winner_val

def get_diagonal_winner(board):
    # check diagonally
    winner_val = None
    board_len = len(board)
    winner_val = board[0][0]
    for i in range(board_len):
        if board[i][i] != winner_val:
            winner_val = None
    if winner_val:
        return winner_val
    winner_val = board[0][board_len - 1]
    for i in range(board_len):
        j = board_len - 1 - i
        if board[i][j] != winner_val:
            winner_val = None
    return winner_val

def winner(board):
    """
    Returns the winner of the game, if there is one.
    """
    winner_val = get_horizontal_winner(board) or get_vertical_winner(board) or get_diagonal_winner(board) or None
    return winner_val

def terminal(board):
    """
    Returns True if game is over, False otherwise.
    """
    if winner(board) != None:
        return True
    for i in board:
        for j in i:
            if j == EMPTY:
                return False
    return True

def utility(board):
    """
    Returns 1 if X has won the game, -1 if O has won, 0 otherwise.
    """
    winner_val = winner(board)
    if winner_val == X:
        return 1
    elif winner_val == O:
        return -1
    return 0

def maxValue(state):
    if terminal(state):
        return utility(state)
    v = -math.inf
    for action in actions(state):
        v = max(v, minValue(result(state, action)))
    return v

def minValue(state):
    if terminal(state):
        return utility(state)
    v = math.inf
    for action in actions(state):
        v = min(v, maxValue(result(state, action)))
    return v

def minimax(board):
    """
    Returns the optimal action for the current player on the board.
    """
    current_player = player(board)
    if current_player == X:
        min = -math.inf
        for action in actions(board):
            check = minValue(result(board, action))  # FIXED
            if check > min:
                min = check
                move = action
    else:
        max = math.inf
        for action in actions(board):
            check = maxValue(result(board, action))  # FIXED
            if check < max:
                max = check
                move = action
    return move

if __name__ == "__main__":
    board = initial_state()
    ai_turn = False
    print("Choose a player")
    user = input()
    if user == "X":
        ai = "O"
    else:
        ai = "X"
    while True:
        game_over = terminal(board)
        playr = player(board)
        if game_over:
            winner = winner(board)
            if winner is None:
                print("Game Over: Tie.")
            else:
                print(f"Game Over: {winner} wins.")
            break;
        else:
            if user != playr and not game_over:
                if ai_turn:
                    move = minimax(board)
                    board = result(board, move)
                    ai_turn = False
                    print(numpy.array(board))
            elif user == playr and not game_over:
                ai_turn = True
                print("Enter the position to move (row,col)")
                i = int(input("Row:"))
                j = int(input("Col:"))
                if board[i][j] == EMPTY:
                    board = result(board, (i, j))
                    print(numpy.array(board))
```

# III. Bài tập ở lớp
Bài 1. Cài đặt thuật toán minimax cho bài toán tictactoe với ma trận 3x3.

```
import math
import copy

X = "X"
O = "O"
EMPTY = None


def initial_state():
    return [[EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY]]

def player(board):
    x_count = sum(row.count(X) for row in board)
    o_count = sum(row.count(O) for row in board)
    return X if x_count == o_count else O

def actions(board):
    return [(i, j) for i in range(3) for j in range(3) if board[i][j] == EMPTY]

def result(board, action):
    new_board = copy.deepcopy(board)
    i, j = action
    new_board[i][j] = player(board)
    return new_board

def winner(board):
    for row in board:
        if row[0] and row[0] == row[1] == row[2]:
            return row[0]
    for col in range(3):
        if board[0][col] and board[0][col] == board[1][col] == board[2][col]:
            return board[0][col]
    if board[0][0] and board[0][0] == board[1][1] == board[2][2]:
        return board[0][0]
    if board[0][2] and board[0][2] == board[1][1] == board[2][0]:
        return board[0][2]
    return None

def terminal(board):
    if winner(board) is not None:
        return True
    return all(board[i][j] != EMPTY for i in range(3) for j in range(3))

def utility(board):
    w = winner(board)
    if w == X: return 1
    if w == O: return -1
    return 0

def max_value(state):

    if terminal(state):
        return utility(state)
    v = -math.inf
    for action in actions(state):
        v = max(v, min_value(result(state, action)))
    return v

def min_value(state):
    if terminal(state):
        return utility(state)
    v = math.inf
    for action in actions(state):
        v = min(v, max_value(result(state, action)))
    return v

def minimax(board):

    if terminal(board):
        return None

    current = player(board)
    best_action = None

    if current == X:
        best_val = -math.inf
        for action in actions(board):
            val = min_value(result(board, action))
            if val > best_val:
                best_val = val
                best_action = action
    else:
        best_val = math.inf
        for action in actions(board):
            val = max_value(result(board, action))
            if val < best_val:
                best_val = val
                best_action = action

    return best_action

def print_board(board):
    print()
    for i, row in enumerate(board):
        print("  " + " | ".join(c if c else "." for c in row))
        if i < 2:
            print("  ---------")
    print()

def print_board_with_index():
    print("\n  Vị trí các ô:")
    for i in range(3):
        print("  " + " | ".join(f"{i},{j}" for j in range(3)))
        if i < 2:
            print("  ---------")
    print()

def main():
    print("=" * 40)
    print("   TICTACTOE 3x3 - MINIMAX ALGORITHM")
    print("=" * 40)

    print_board_with_index()

    human = ""
    while human not in [X, O]:
        human = input("Bạn chọn X hay O? ").strip().upper()
    ai = O if human == X else X
    print(f"\n>> Bạn: {human}  |  AI: {ai}")
    print(">> X luôn đi trước\n")

    board = initial_state()

    while not terminal(board):
        print_board(board)
        current = player(board)

        if current == human:
            print(f"Lượt của bạn ({human}):")
            while True:
                try:
                    r = int(input("  Nhập hàng (0-2): "))
                    c = int(input("  Nhập cột (0-2): "))
                    if 0 <= r <= 2 and 0 <= c <= 2 and board[r][c] == EMPTY:
                        board = result(board, (r, c))
                        break
                    else:
                        print("  >> Ô không hợp lệ hoặc đã có quân! Thử lại.")
                except ValueError:
                    print("  >> Vui lòng nhập số nguyên!")
        else:
            print(f"Lượt của AI ({ai}) - đang tính toán...")
            move = minimax(board)
            board = result(board, move)
            print(f"  >> AI đánh vào: hàng {move[0]}, cột {move[1]}")

    print_board(board)
    w = winner(board)
    print("=" * 40)
    if w == human:
        print("CHÚC MỪNG! Bạn thắng!")
    elif w == ai:
        print("AI thắng! Chúc bạn may mắn lần sau.")
    else:
        print("Hòa! Trận đấu cân bằng.")
    print("=" * 40)

if __name__ == "__main__":
    main()
```

# IV. Bài tập về nhà
Bài 3. Cài đặt thuật toán minimax cho bài toán tictactoe với ma trận 5x5, 10x10, NxN.

```
import math
import copy

X = "X"
O = "O"
EMPTY = None

def get_win_length(n):

    if n <= 4:
        return n
    return 5

def get_max_depth(n):

    if n <= 3:  return 9
    if n <= 5:  return 4
    if n <= 8:  return 3
    return 2


def initial_state(n):
    return [[EMPTY] * n for _ in range(n)]

def player(board):
    n = len(board)
    x_count = sum(row.count(X) for row in board)
    o_count = sum(row.count(O) for row in board)
    return X if x_count == o_count else O

def actions(board):
    n = len(board)
    return [(i, j) for i in range(n) for j in range(n) if board[i][j] == EMPTY]

def result(board, action):
    new_board = copy.deepcopy(board)
    i, j = action
    new_board[i][j] = player(board)
    return new_board

def check_winner(board):

    n = len(board)
    win_len = get_win_length(n)

    for i in range(n):
        for j in range(n - win_len + 1):
            seq = [board[i][j + k] for k in range(win_len)]
            if seq[0] and all(c == seq[0] for c in seq):
                return seq[0]

    for j in range(n):
        for i in range(n - win_len + 1):
            seq = [board[i + k][j] for k in range(win_len)]
            if seq[0] and all(c == seq[0] for c in seq):
                return seq[0]

    for i in range(n - win_len + 1):
        for j in range(n - win_len + 1):
            seq = [board[i + k][j + k] for k in range(win_len)]
            if seq[0] and all(c == seq[0] for c in seq):
                return seq[0]

    for i in range(n - win_len + 1):
        for j in range(win_len - 1, n):
            seq = [board[i + k][j - k] for k in range(win_len)]
            if seq[0] and all(c == seq[0] for c in seq):
                return seq[0]
    return None

def terminal(board):
    if check_winner(board) is not None:
        return True
    return all(board[i][j] != EMPTY for i in range(len(board)) for j in range(len(board)))

def score_sequence(seq, win_len, piece, opp):

    if opp in seq:
        return 0

    count = seq.count(piece)
    if count == 0:
        return 0

    scores = {1: 1, 2: 10, 3: 100, 4: 1000}
    return scores.get(count, 10000)

def heuristic(board, maximizing_piece):

    n = len(board)
    win_len = get_win_length(n)
    opp = O if maximizing_piece == X else X
    score = 0

    def evaluate_line(line):
        total = 0
        for start in range(len(line) - win_len + 1):
            seq = line[start:start + win_len]
            total += score_sequence(seq, win_len, maximizing_piece, opp)
            total -= score_sequence(seq, win_len, opp, maximizing_piece)
        return total

    for i in range(n):
        score += evaluate_line(board[i])

    for j in range(n):
        col = [board[i][j] for i in range(n)]
        score += evaluate_line(col)

    for i in range(n - win_len + 1):
        for j in range(n - win_len + 1):
            diag = [board[i + k][j + k] for k in range(win_len)]
            score += evaluate_line(diag)

    for i in range(n - win_len + 1):
        for j in range(win_len - 1, n):
            diag = [board[i + k][j - k] for k in range(win_len)]
            score += evaluate_line(diag)

    return score

INF = math.inf

def minimax(board, depth, max_depth, is_maximizing, maximizing_piece):
    w = check_winner(board)
    if w == maximizing_piece:
        return 10000 - depth
    if w is not None:
        return -10000 + depth
    if terminal(board):
        return 0
    if depth >= max_depth:
        return heuristic(board, maximizing_piece)

    if is_maximizing:
        best = -INF
        for action in actions(board):
            val = minimax(result(board, action), depth + 1, max_depth, False, maximizing_piece)
            best = max(best, val)
        return best
    else:
        best = INF
        for action in actions(board):
            val = minimax(result(board, action), depth + 1, max_depth, True, maximizing_piece)
            best = min(best, val)
        return best

def find_best_move(board, ai_piece):
    n = len(board)
    max_depth = get_max_depth(n)
    is_max = (ai_piece == X)

    best_action = None
    best_val = -INF if is_max else INF

    for action in actions(board):
        val = minimax(result(board, action), 0, max_depth, not is_max, ai_piece if is_max else (O if ai_piece == X else X))
        val = minimax(result(board, action), 0, max_depth, False, ai_piece)
        if val > best_val:
            best_val = val
            best_action = action

    return best_action


def print_board(board):
    n = len(board)
    col_width = 2 if n < 10 else 3
    header = "   " + " ".join(str(j).rjust(col_width) for j in range(n))
    print(header)
    print("   " + "-" * (col_width * n + n - 1))
    for i, row in enumerate(board):
        cells = " ".join((c if c else ".").rjust(col_width) for c in row)
        print(f"{i:2d} | {cells}")
    print()

def main():
    print("=" * 50)
    print("    TICTACTOE NxN - MINIMAX (DEPTH-LIMITED)")
    print("=" * 50)

    try:
        n = int(input("Nhập kích thước bảng N (vd: 3, 5, 10): "))
        if n < 3:
            n = 3
            print("Đặt lại N = 3 (tối thiểu)")
    except ValueError:
        n = 3

    win_len = get_win_length(n)
    max_depth = get_max_depth(n)
    print(f"\n>> Bảng: {n}x{n} | Điều kiện thắng: {win_len} quân liên tiếp | Độ sâu tìm kiếm: {max_depth}")

    human = ""
    while human not in [X, O]:
        human = input("Bạn chọn X hay O? ").strip().upper()
    ai = O if human == X else X
    print(f">> Bạn: {human}  |  AI: {ai}  (X đi trước)\n")

    board = initial_state(n)

    while not terminal(board):
        print_board(board)
        current = player(board)

        if current == human:
            print(f"Lượt của bạn ({human}):")
            while True:
                try:
                    r = int(input(f"  Hàng (0-{n-1}): "))
                    c = int(input(f"  Cột (0-{n-1}): "))
                    if 0 <= r < n and 0 <= c < n and board[r][c] == EMPTY:
                        board = result(board, (r, c))
                        break
                    else:
                        print("  >> Ô không hợp lệ! Thử lại.")
                except ValueError:
                    print("  >> Vui lòng nhập số nguyên!")
        else:
            print(f"Lượt của AI ({ai}) - đang tính toán (Minimax độ sâu {max_depth})...")
            move = find_best_move(board, ai)
            if move:
                board = result(board, move)
                print(f"  >> AI đánh: hàng {move[0]}, cột {move[1]}")
            else:
                print("  >> Không tìm được nước đi!")
                break

    print_board(board)
    w = check_winner(board)
    print("=" * 50)
    if w == human:
        print("CHÚC MỪNG! Bạn thắng!")
    elif w == ai:
        print("AI thắng!")
    else:
        print("Hòa!")
    print("=" * 50)

if __name__ == "__main__":
    main()
```
