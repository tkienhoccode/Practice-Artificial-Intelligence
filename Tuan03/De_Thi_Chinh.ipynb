````python
# CÂU 1a – Sao chép code từ ảnh, sửa 4 lỗi và giải thích bằng comment
# =====================================================================

# LỖI 1 (hàm tinh_phi): Điều kiện sai – dùng '<' thay vì '>' 
#   Phụ phí chỉ áp dụng khi trọng lượng > 5kg, không phải < 5kg
#   Sửa: if self.trong_luong_kg < 5  →  if self.trong_luong_kg > 5

# LỖI 2 (hàm tinh_phi): Trả về phi_co_ban - phu_phi (trừ thay vì cộng)
#   Phụ phí phải được cộng thêm vào phí cơ bản, không phải trừ đi
#   Sửa: return phi_co_ban - phu_phi  →  return phi_co_ban + phu_phi

# LỖI 3 (hàm tong_doanh_thu): Dùng '=' thay vì '+=' nên tong luôn bị ghi đè
#   Mỗi vòng lặp cần cộng dồn, không phải gán lại
#   Sửa: tong = don.tinh_phi()  →  tong += don.tinh_phi()

# LỖI 4 (hàm loc_don_xa): Dùng '=' (gán) thay vì '>=' (so sánh) trong list comprehension
#   Điều kiện lọc cần toán tử so sánh, không phải gán
#   Sửa: if don.khoang_cach_km = nguong_km  →  if don.khoang_cach_km >= nguong_km

class DonHang:

    def __init__(self, ma_don, nguoi_nhan, khoang_cach_km, trong_luong_kg):
        self.ma_don = ma_don
        self.nguoi_nhan = nguoi_nhan
        self.khoang_cach_km = khoang_cach_km
        self.trong_luong_kg = trong_luong_kg
        self.trang_thai = 'cho_lay_hang'

    def tinh_phi(self):
        # Phí cơ bản: 15,000đ/km; phụ phí nặng: 5,000đ/kg nếu > 5kg
        phi_co_ban = self.khoang_cach_km * 15000
        phu_phi = 0

        if self.trong_luong_kg > 5:           # ĐÃ SỬA LỖI 1: > thay vì <
            phu_phi = self.trong_luong_kg * 5000

        return phi_co_ban + phu_phi           # ĐÃ SỬA LỖI 2: + thay vì -

    def __str__(self):
        return (f"Đơn {self.ma_don} | {self.nguoi_nhan} | "
                f"{self.khoang_cach_km}km | "
                f"Phí: {self.tinh_phi():,}đ | {self.trang_thai}")


def tong_doanh_thu(danh_sach_don):
    tong = 0

    for don in danh_sach_don:
        tong += don.tinh_phi()                # ĐÃ SỬA LỖI 3: += thay vì =

    return tong


def loc_don_xa(danh_sach_don, nguong_km):
    # Lọc đơn hàng có khoảng cách >= nguong_km
    return [don for don in danh_sach_don
            if don.khoang_cach_km >= nguong_km]  # ĐÃ SỬA LỖI 4: >= thay vì =


# Kiểm thử
d1 = DonHang('DH001', 'Nguyen Van A', 10, 3)
d2 = DonHang('DH002', 'Tran Thi B', 20, 8)
d3 = DonHang('DH003', 'Le Van C', 5, 2)

ds = [d1, d2, d3]
for d in ds:
    print(d)

print(f"\nTổng doanh thu: {tong_doanh_thu(ds):,}đ")
print("\nĐơn xa >= 10km:")
for d in loc_don_xa(ds, 10):
    print(" ", d)
````

```

Đơn xa >= 10km:
  Đơn DH001 | Nguyen Van A | 10km | Phí: 150,000đ | cho_lay_hang
  Đơn DH002 | Tran Thi B | 20km | Phí: 340,000đ | cho_lay_hang
```

**Giải thích:**
- DH001: 10km × 15,000 = 150,000đ (3kg ≤ 5kg → không có phụ phí)
- DH002: 20km × 15,000 + 8kg × 5,000 = 300,000 + 40,000 = 340,000đ (8kg > 5kg → có phụ phí)
- DH003: 5km × 15,000 = 75,000đ (2kg ≤ 5kg → không có phụ phí)
- Tổng: 150,000 + 340,000 + 75,000 = 565,000đ

````python
# CÂU 1c – Thêm phương thức cap_nhat_trang_thai
# Thứ tự hợp lệ: 'cho_lay_hang' -> 'dang_van_chuyen' -> 'da_giao'

class DonHang:

    LUONG_TRANG_THAI = ['cho_lay_hang', 'dang_van_chuyen', 'da_giao']

    def __init__(self, ma_don, nguoi_nhan, khoang_cach_km, trong_luong_kg):
        self.ma_don = ma_don
        self.nguoi_nhan = nguoi_nhan
        self.khoang_cach_km = khoang_cach_km
        self.trong_luong_kg = trong_luong_kg
        self.trang_thai = 'cho_lay_hang'

    def tinh_phi(self):
        phi_co_ban = self.khoang_cach_km * 15000
        phu_phi = 0
        if self.trong_luong_kg > 5:
            phu_phi = self.trong_luong_kg * 5000
        return phi_co_ban + phu_phi

    def cap_nhat_trang_thai(self, trang_thai_moi):
        """Chỉ cho phép chuyển đúng thứ tự: cho_lay_hang -> dang_van_chuyen -> da_giao"""
        vi_tri_hien_tai = self.LUONG_TRANG_THAI.index(self.trang_thai)
        if trang_thai_moi not in self.LUONG_TRANG_THAI:
            print(f"Lỗi: Trạng thái '{trang_thai_moi}' không hợp lệ.")
            return
        vi_tri_moi = self.LUONG_TRANG_THAI.index(trang_thai_moi)
        if vi_tri_moi == vi_tri_hien_tai + 1:
            self.trang_thai = trang_thai_moi
            print(f"Đơn {self.ma_don}: Cập nhật thành công → {trang_thai_moi}")
        else:
            print(f"Lỗi: Không thể chuyển từ '{self.trang_thai}' sang '{trang_thai_moi}'. Sai thứ tự!")

    def __str__(self):
        return (f"Đơn {self.ma_don} | {self.nguoi_nhan} | "
                f"{self.khoang_cach_km}km | "
                f"Phí: {self.tinh_phi():,}đ | {self.trang_thai}")


# Kiểm thử
d = DonHang('DH001', 'Nguyen Van A', 10, 3)
print(d)

d.cap_nhat_trang_thai('dang_van_chuyen')   # Đúng thứ tự → thành công
print(d)

d.cap_nhat_trang_thai('cho_lay_hang')      # Sai thứ tự → báo lỗi

d.cap_nhat_trang_thai('da_giao')           # Đúng thứ tự → thành công
print(d)

d.cap_nhat_trang_thai('dang_van_chuyen')   # Sai thứ tự (đã giao rồi) → báo lỗi
````

**Câu 2a – Bảng trace A* (điền tay):**

| Bước | Node xét | g(n) | h(n) | f(n) | OPEN (node:f) sau bước này | CLOSE |
|------|----------|------|------|------|-----------------------------|-------|
| 1    | S        | 0    | 6    | 6    | A:6, B:6                    | {S}   |
| 2    | A        | 1    | 5    | 6    | B:6, C:12, D:10             | {S,A} |
| 3    | B        | 3    | 3    | 6    | E:6, C:12, D:10             | {S,A,B} |
| 4    | E        | 5    | 1    | 6    | G:6, C:12, D:10             | {S,A,B,E} |
| 5    | G        | 6    | 0    | 6    | —                           | {S,A,B,E,G} |

**Đường đi tìm được:** S → B → E → G  
**Tổng chi phí:** 6

---

**Câu 2a – What-if (0.25đ bonus):**

> Nếu thay `h(E) = 6` thay vì `h(E) = 1`, thuật toán A* có chọn cùng đường đi không?

Khi đang xét node B (g=3), mở E với g(E)=5, h(E)=6 → **f(E)=11**.
Lúc này OPEN có: E:11, C:12, D:10.
A* sẽ chọn **D** (f=10) thay vì E. Từ D → G có chi phí 4 → tổng g(G) = 1+5+4 = 10.
Kết quả: A* chọn đường **S → A → D → G** (chi phí 10), **không còn chọn S → B → E → G** nữa.

````python
# CÂU 2b – Cài đặt A*
import heapq

graph = {
    'S': [('A', 1), ('B', 3)],
    'A': [('C', 2), ('D', 5)],
    'B': [('E', 2)],
    'C': [],
    'D': [('G', 4)],
    'E': [('G', 1)],
    'G': []
}

heuristic = {
    'S': 6,
    'A': 5,
    'B': 3,
    'C': 9,
    'D': 4,
    'E': 1,
    'G': 0
}


def astar(graph, heuristic, start, goal):
    """
    Thuật toán A*.
    In từng bước theo định dạng:
      Buoc X | Xet: <node> | g=? h=? f=? | OPEN: [<node>:f, ...]
    Trả về (đường_đi, tổng_chi_phí).
    """
    # (f, g, node, path)
    open_list = []
    heapq.heappush(open_list, (heuristic[start], 0, start, [start]))
    closed = set()
    step = 0

    while open_list:
        f, g, node, path = heapq.heappop(open_list)

        if node in closed:
            continue

        closed.add(node)
        step += 1
        h = heuristic[node]

        open_display = ', '.join(f"{n}:{fn}" for fn, _, n, _ in sorted(open_list))
        print(f"Buoc {step} | Xet: {node} | g={g} h={h} f={f} | OPEN: [{open_display}]")

        if node == goal:
            return path, g

        for neighbor, cost in graph.get(node, []):
            if neighbor not in closed:
                new_g = g + cost
                new_f = new_g + heuristic[neighbor]
                heapq.heappush(open_list, (new_f, new_g, neighbor, path + [neighbor]))

    return None, float('inf')


print("=" * 55)
print("A* – Tim duong tu S den G")
print("=" * 55)
path, cost = astar(graph, heuristic, 'S', 'G')
if path:
    print(f"\nDuong di: {' -> '.join(path)}")
    print(f"Tong chi phi: {cost}")
````

**Câu 2c – Kiểm tra admissible:**

- h*(S) = chi phí thực từ S đến G = S→B→E→G = 3+2+1 = **6**  →  h(S)=6  →  6 ≤ 6 ✓
- h*(A) = chi phí thực từ A đến G = A→D→G = 5+4 = **9**  →  h(A)=5  →  5 ≤ 9 ✓
- h*(B) = chi phí thực từ B đến G = B→E→G = 2+1 = **3**  →  h(B)=3  →  3 ≤ 3 ✓
- h*(E) = chi phí thực từ E đến G = E→G = **1**  →  h(E)=1  →  1 ≤ 1 ✓
- h*(D) = chi phí thực từ D đến G = D→G = **4**  →  h(D)=4  →  4 ≤ 4 ✓

**Kết luận:** Heuristic h(n) **admissible** vì với mọi node đã kiểm tra, h(n) ≤ h*(n). Heuristic không bao giờ ước lượng cao hơn chi phí thực tế đến đích, đảm bảo A* luôn tìm được đường đi tối ưu.

# CÂU 3 (2.5 ĐIỂM) – Điền Hàm Minimax

**a) (1.5đ)** Quan sát code bên dưới. Điền vào **7 chỗ trống** `___(n)`. Mỗi chỗ là một biểu thức hoặc tên hàm ngắn. **Không thêm hoặc xóa dòng.**

**b) (0.5đ)** Cho trạng thái:
```
X | O | X
---------
O | X | O
---------
  |   |  
```
Đến lượt ai? Minimax gợi ý nước đi nào?

**c) (0.5đ)** Cài đặt `minimax(board, is_maximizing)` và `best_move(board)`, kiểm thử với trạng thái trên.

````python
from IPython.display import Image, display
display(Image('cau3_code.png', width=950))
````

````python
# CÂU 3a – Điền 7 chỗ trống
import copy, math

X, O, EMPTY = 'X', 'O', None

def initial_state():
    return [[EMPTY]*3 for _ in range(3)]

def player(board):
    x_count = sum(row.count(X) for row in board)
    o_count = sum(row.count(O) for row in board)
    # TRỐNG 1: X đi trước (khi số X bằng số O)
    # TRỐNG 2: O đi khi X nhiều hơn O
    return X if x_count == o_count else O   # __(1)=X, __(2)=O

def actions(board):
    return [(i, j) for i in range(3) for j in range(3)
            if board[i][j] == EMPTY]         # __(3)=EMPTY

def result(board, action):
    i, j = action
    if board[i][j] is not EMPTY:
        raise ValueError("Ô đã đánh!")
    new_board = copy.deepcopy(board)
    new_board[i][j] = player(board)          # __(4)=player
    return new_board

def winner(board):
    lines = []
    for i in range(3):
        lines.append(board[i])
        lines.append([board[r][i] for r in range(3)])
    lines.append([board[i][i] for i in range(3)])
    lines.append([board[i][2-i] for i in range(3)])
    for line in lines:
        if line[0] is not EMPTY and line[0] == line[1] == line[2]:  # __(5)=line[0] == line[1] == line[2]
            return line[0]
    return None

def terminal(board):
    if winner(board) is not None:
        return True
    return all(board[i][j] is not EMPTY for i in range(3) for j in range(3))

def utility(board):
    w = winner(board)
    if w == X:
        return 1
    elif w == O:
        return -1
    return 0

def max_value(state):
    if terminal(state):
        return utility(state)
    v = -math.inf
    for action in actions(state):
        v = max(v, min_value(result(state, action)))  # __(6)=min_value
    return v

def min_value(state):
    if terminal(state):
        return utility(state)
    v = math.inf
    for action in actions(state):
        v = min(v, max_value(result(state, action)))  # __(7)=max_value
    return v


print("7 chỗ trống đã điền:")
print("  (1) X")
print("  (2) O")
print("  (3) EMPTY")
print("  (4) player")
print("  (5) line[0] == line[1] == line[2]")
print("  (6) min_value")
print("  (7) max_value")
````

**Câu 3b – Phân tích trạng thái:**

```
X | O | X
---------
O | X | O
---------
  |   |  
```

- Đếm: X có 3 quân, O có 3 quân → **đến lượt X** (x_count == o_count → return X)
- X cần chặn: O không có nguy cơ thắng ngay ở hàng 3. X nên đánh vào **(2,2)** (góc dưới phải) để tạo nguy cơ thắng chéo.
- Minimax sẽ gợi ý nước đi tối ưu cho X trong số các ô trống: (2,0), (2,1), (2,2).

````python
# CÂU 3c – Cài đặt minimax và best_move, kiểm thử

def minimax(board, is_maximizing):
    """Minimax thuần."""
    if terminal(board):
        return utility(board)
    if is_maximizing:
        v = -math.inf
        for action in actions(board):
            v = max(v, minimax(result(board, action), False))
        return v
    else:
        v = math.inf
        for action in actions(board):
            v = min(v, minimax(result(board, action), True))
        return v


def best_move(board):
    """Trả về (i,j) – nước đi tốt nhất cho người chơi hiện tại."""
    current = player(board)
    best_val = -math.inf if current == X else math.inf
    best_action = None

    for action in actions(board):
        val = minimax(result(board, action), current != X)
        if current == X and val > best_val:
            best_val = val
            best_action = action
        elif current == O and val < best_val:
            best_val = val
            best_action = action

    return best_action


# Kiểm thử với trạng thái đề bài
board_test = [
    [X, O, X],
    [O, X, O],
    [EMPTY, EMPTY, EMPTY]
]

print("Trạng thái hiện tại:")
for row in board_test:
    print([v if v else '.' for v in row])

print(f"\nĐến lượt: {player(board_test)}")
move = best_move(board_test)
print(f"Nước đi tốt nhất (Minimax): {move}")
````

## Phần B – Cài đặt (2.0đ)

*(Dùng lại `initial_state`, `player`, `actions`, `result`, `winner`, `terminal`, `utility` từ Câu 3 — chạy cell Câu 3 trước)*

**b1) (1.0đ)** Cài đặt `minimax_ab(board, depth, alpha, beta, is_maximizing)`:
- Điểm: `10 - depth` (X thắng), `depth - 10` (O thắng), `0` (hòa)
- Cắt tỉa khi `beta <= alpha`

**b2) (0.5đ)** Cài đặt `find_best_move(board)` trả về nước đi tốt nhất.

**b3) (0.5đ)** Đếm và so sánh số lần gọi đệ quy của Minimax thuần (Câu 3) và Alpha-Beta (Câu 4) từ cùng một bảng khởi đầu. Nhận xét.

````python
# CÂU 4 – Alpha-Beta Pruning
# (Chạy cell Câu 3 trước để có các hàm hỗ trợ)
import math

call_ab = 0
call_mm = 0


# b1) Minimax + Alpha-Beta Pruning
def minimax_ab(board, depth, alpha, beta, is_maximizing):
    """
    Minimax + Alpha-Beta Pruning.
    Diem: (10-depth) neu X thang, (depth-10) neu O thang, 0 neu hoa.
    Cat tinh khi beta <= alpha.
    """
    global call_ab
    call_ab += 1

    # Kiem tra terminal state va tinh diem
    if terminal(board):
        w = winner(board)
        if w == X:
            return 10 - depth
        elif w == O:
            return depth - 10
        return 0

    if is_maximizing:
        best = -math.inf
        for action in actions(board):
            val = minimax_ab(result(board, action), depth + 1, alpha, beta, False)
            best = max(best, val)
            alpha = max(alpha, best)
            if beta <= alpha:   # Cat tinh
                break
        return best
    else:
        best = math.inf
        for action in actions(board):
            val = minimax_ab(result(board, action), depth + 1, alpha, beta, True)
            best = min(best, val)
            beta = min(beta, best)
            if beta <= alpha:   # Cat tinh
                break
        return best


# b2) Tim nuoc di tot nhat
def find_best_move(board):
    """Tra ve (i,j) – nuoc di tot nhat cho nguoi choi hien tai."""
    current = player(board)
    best_val = -math.inf if current == X else math.inf
    best_action = None

    for action in actions(board):
        val = minimax_ab(result(board, action), 1, -math.inf, math.inf, current != X)
        if current == X and val > best_val:
            best_val = val
            best_action = action
        elif current == O and val < best_val:
            best_val = val
            best_action = action

    return best_action


# b3) Minimax thuan de so sanh
def minimax_count(board, is_maximizing):
    """Minimax thuan co dem so lan goi."""
    global call_mm
    call_mm += 1

    if terminal(board):
        return utility(board)

    if is_maximizing:
        v = -math.inf
        for action in actions(board):
            v = max(v, minimax_count(result(board, action), False))
        return v
    else:
        v = math.inf
        for action in actions(board):
            v = min(v, minimax_count(result(board, action), True))
        return v


# Thuc nghiem so sanh
board0 = initial_state()

call_mm = 0
minimax_count(board0, True)

call_ab = 0
minimax_ab(board0, 0, -math.inf, math.inf, True)

print("=" * 50)
print("SO SANH MINIMAX vs ALPHA-BETA")
print("=" * 50)
print(f"Minimax thuan tuy  : {call_mm:>7,} lan goi")
print(f"Alpha-Beta Pruning : {call_ab:>7,} lan goi")
if call_mm > 0:
    print(f"Tiet kiem          : {(1 - call_ab/call_mm)*100:.1f}%")
````
