# II. Bài tập hướng dẫn mẫu
b.	Thuật toán cắt tỉa Alpha-beta với ứng dụng vào bài toán TicTacToe

```
import os, math

def GetWinner(board):
    """
    Returns the winner in the current board if there is one, otherwise it returns None.
    """
    # horizontal
    if board[0] == board[1] and board[1] == board[2]:
        return board[0]
    elif board[3] == board[4] and board[4] == board[5]:
        return board[3]
    elif board[6] == board[7] and board[7] == board[8]:
        return board[6]
    # vertical
    elif board[0] == board[3] and board[3] == board[6]:
        return board[0]
    elif board[1] == board[4] and board[4] == board[7]:
        return board[1]
    elif board[2] == board[5] and board[5] == board[8]:
        return board[2]
    # diagonal
    elif board[0] == board[4] and board[4] == board[8]:
        return board[0]
    elif board[2] == board[4] and board[4] == board[6]:
        return board[2]

def PrintBoard(board):
    """
    Clears the console and prints the current board.
    """
    os.system('cls' if os.name=='nt' else 'clear')
    print(f'''
    {board[0]}|{board[1]}|{board[2]}
    {board[3]}|{board[4]}|{board[5]}
    {board[6]}|{board[7]}|{board[8]}
    ''')

def GetAvailableCells(board):
    """
    Returns a list of indices containing all available cells in a board.
    """
    available = list()
    for cell in board:
        if cell != "X" and cell != "O":
            available.append(cell)
    return available

def minimax(position, depth, alpha, beta, isMaximizing):
    winner = GetWinner(position)
    if winner != None:
        return 10 - depth if winner == "X" else -10 + depth
    if len(GetAvailableCells(position)) == 0:
        return 0

    if isMaximizing:
        maxEval = -math.inf
        for cell in GetAvailableCells(position):
            position[cell - 1] = "X"
            Eval = minimax(position, depth + 1, alpha, beta, False)
            maxEval = max(maxEval, Eval)
            alpha = max(alpha, Eval)
            position[cell - 1] = cell
            if beta <= alpha:
                break # prune
        return maxEval
    else:
        minEval = +math.inf
        for cell in GetAvailableCells(position):
            position[cell - 1] = "O"
            Eval = minimax(position, depth + 1, alpha, beta, True)
            minEval = min(minEval, Eval)
            beta = min(beta, Eval)
            position[cell - 1] = cell
            if beta <= alpha:
                break # prune
        return minEval

def FindBestMove(currentPosition, AI):
    bestVal = -math.inf if AI == "X" else +math.inf
    bestMove = -1
    for cell in GetAvailableCells(currentPosition):
        currentPosition[cell - 1] = AI
        moveVal = minimax(currentPosition, 0, -math.inf, +math.inf, False if AI == "X" else True)
        currentPosition[cell - 1] = cell
        if (AI == "X" and moveVal > bestVal):
            bestMove = cell
            bestVal = moveVal
        elif (AI == "O" and moveVal < bestVal):
            bestMove = cell
            bestVal = moveVal
    return bestMove

def main():
    player = input("Play as X or O? ").strip().upper()
    AI = "O" if player == "X" else "X"
    currentGame = [*range(1, 10)]
    # X always starts first.
    currentTurn = "X"
    counter = 0
    while True:
        if currentTurn == AI:
            # NOTE: if the AI starts first, it'll always choose index 0 so to save time you could play it.
            cell = FindBestMove(currentGame, AI)
            currentGame[cell - 1] = AI
            currentTurn = player
        elif currentTurn == player:
            PrintBoard(currentGame)
            while True:
                humanInput = int(input("Enter Number: ").strip())
                if humanInput in currentGame:
                    currentGame[humanInput - 1] = player
                    currentTurn = AI
                    break
                else:
                    PrintBoard(currentGame)
                    print("Cell Not Available.")
        if GetWinner(currentGame) != None:
            PrintBoard(currentGame)
            print(f"{GetWinner(currentGame)} WON!!!")
            break
        counter += 1
        if GetWinner(currentGame) == None and counter == 9:
            PrintBoard(currentGame)
            print("Tie.")
            break

if __name__ == "__main__":
    main()
```

# III. Bài tập ở lớp
Bài 2. Cài đặt thuật toán cắt tỉa Alpha-beta với ứng dụng vào bài toán TicTacToe với ma trận 3x3.

```
import math
import copy

X = "X"
O = "O"
EMPTY = None


def initial_state():
    return [[EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY]]

def player(board):
    x_count = sum(row.count(X) for row in board)
    o_count = sum(row.count(O) for row in board)
    return X if x_count == o_count else O

def actions(board):
    return [(i, j) for i in range(3) for j in range(3) if board[i][j] == EMPTY]

def result(board, action):
    new_board = copy.deepcopy(board)
    i, j = action
    new_board[i][j] = player(board)
    return new_board

def winner(board):
    for row in board:
        if row[0] and row[0] == row[1] == row[2]:
            return row[0]
    for col in range(3):
        if board[0][col] and board[0][col] == board[1][col] == board[2][col]:
            return board[0][col]
    if board[0][0] and board[0][0] == board[1][1] == board[2][2]:
        return board[0][0]
    if board[0][2] and board[0][2] == board[1][1] == board[2][0]:
        return board[0][2]
    return None

def terminal(board):
    if winner(board) is not None:
        return True
    return all(board[i][j] != EMPTY for i in range(3) for j in range(3))

def utility(board):
    w = winner(board)
    if w == X: return 1
    if w == O: return -1
    return 0

nodes_visited = 0

def alpha_beta(state, alpha, beta, is_maximizing):

    global nodes_visited
    nodes_visited += 1

    if terminal(state):
        return utility(state)

    if is_maximizing:
        max_eval = -math.inf
        for action in actions(state):
            eval_val = alpha_beta(result(state, action), alpha, beta, False)
            max_eval = max(max_eval, eval_val)
            alpha = max(alpha, eval_val)

            if beta <= alpha:
                break
        return max_eval
    else:
        min_eval = math.inf
        for action in actions(state):
            eval_val = alpha_beta(result(state, action), alpha, beta, True)
            min_eval = min(min_eval, eval_val)
            beta = min(beta, eval_val)

            if beta <= alpha:
                break
        return min_eval

def find_best_move(board):

    global nodes_visited
    nodes_visited = 0

    current = player(board)
    best_action = None

    if current == X:
        best_val = -math.inf
        for action in actions(board):
            val = alpha_beta(result(board, action), -math.inf, math.inf, False)
            if val > best_val:
                best_val = val
                best_action = action
    else:
        best_val = math.inf
        for action in actions(board):
            val = alpha_beta(result(board, action), -math.inf, math.inf, True)
            if val < best_val:
                best_val = val
                best_action = action

    return best_action, best_val, nodes_visited


def print_board(board):
    print()
    for i, row in enumerate(board):
        print("  " + " | ".join(c if c else "." for c in row))
        if i < 2:
            print("  ---------")
    print()

def print_board_with_index():
    print("\n  Vị trí các ô:")
    for i in range(3):
        print("  " + " | ".join(f"{i},{j}" for j in range(3)))
        if i < 2:
            print("  ---------")
    print()

def main():
    print("=" * 50)
    print("  TICTACTOE 3x3 - ALPHA-BETA PRUNING ALGORITHM")
    print("=" * 50)

    print_board_with_index()

    human = ""
    while human not in [X, O]:
        human = input("Bạn chọn X hay O? ").strip().upper()
    ai = O if human == X else X
    print(f"\n>> Bạn: {human}  |  AI: {ai}")
    print(">> X luôn đi trước\n")

    board = initial_state()
    total_nodes = 0

    while not terminal(board):
        print_board(board)
        current = player(board)

        if current == human:
            print(f"Lượt của bạn ({human}):")
            while True:
                try:
                    r = int(input("  Hàng (0-2): "))
                    c = int(input("  Cột (0-2): "))
                    if 0 <= r <= 2 and 0 <= c <= 2 and board[r][c] == EMPTY:
                        board = result(board, (r, c))
                        break
                    else:
                        print("  >> Ô không hợp lệ! Thử lại.")
                except ValueError:
                    print("  >> Vui lòng nhập số nguyên!")
        else:
            print(f"Lượt của AI ({ai}) - đang tính toán với Alpha-Beta...")
            move, score, visited = find_best_move(board)
            total_nodes += visited
            board = result(board, move)
            print(f"  >> AI đánh: hàng {move[0]}, cột {move[1]}")
            print(f"  >> Điểm đánh giá: {score} | Nút đã duyệt: {visited}")

    print_board(board)
    w = winner(board)
    print("=" * 50)
    if w == human:
        print("CHÚC MỪNG! Bạn thắng!")
    elif w == ai:
        print("AI thắng!")
    else:
        print("Hòa!")
    print(f"  Tổng số nút AI đã duyệt: {total_nodes}")
    print("=" * 50)

if __name__ == "__main__":
    main()
```

# IV. Bài tập về nhà
Bài 4. Cài đặt thuật toán cắt tỉa Alpha-beta cho bài toán tictactoe với ma trận 5x5, 10x10, NxN.

```
import math
import copy

X = "X"
O = "O"
EMPTY = None


def get_win_length(n):
    return n if n <= 4 else 5

def get_max_depth(n):
    if n <= 3:  return 9
    if n <= 5:  return 5
    if n <= 8:  return 4
    return 3


def initial_state(n):
    return [[EMPTY] * n for _ in range(n)]

def player(board):
    x_count = sum(row.count(X) for row in board)
    o_count = sum(row.count(O) for row in board)
    return X if x_count == o_count else O

def actions_ordered(board):

    n = len(board)
    center = n / 2
    moves = [(i, j) for i in range(n) for j in range(n) if board[i][j] == EMPTY]

    moves.sort(key=lambda m: abs(m[0] - center) + abs(m[1] - center))
    return moves

def result(board, action):
    new_board = copy.deepcopy(board)
    i, j = action
    new_board[i][j] = player(board)
    return new_board

def check_winner(board):
    n = len(board)
    win_len = get_win_length(n)

    for i in range(n):
        for j in range(n - win_len + 1):
            seq = [board[i][j + k] for k in range(win_len)]
            if seq[0] and all(c == seq[0] for c in seq):
                return seq[0]
    for j in range(n):
        for i in range(n - win_len + 1):
            seq = [board[i + k][j] for k in range(win_len)]
            if seq[0] and all(c == seq[0] for c in seq):
                return seq[0]
    for i in range(n - win_len + 1):
        for j in range(n - win_len + 1):
            seq = [board[i + k][j + k] for k in range(win_len)]
            if seq[0] and all(c == seq[0] for c in seq):
                return seq[0]
    for i in range(n - win_len + 1):
        for j in range(win_len - 1, n):
            seq = [board[i + k][j - k] for k in range(win_len)]
            if seq[0] and all(c == seq[0] for c in seq):
                return seq[0]
    return None

def terminal(board):
    if check_winner(board) is not None:
        return True
    return all(board[i][j] != EMPTY for i in range(len(board)) for j in range(len(board)))


def score_sequence(seq, win_len, piece, opp):
    if opp in seq:
        return 0
    count = seq.count(piece)
    if count == 0:
        return 0
    scores = {1: 1, 2: 10, 3: 100, 4: 1000}
    return scores.get(count, 10000)

def heuristic(board, ai_piece):
    n = len(board)
    win_len = get_win_length(n)
    opp = O if ai_piece == X else X
    score = 0

    def eval_line(line):
        total = 0
        for s in range(len(line) - win_len + 1):
            seq = line[s:s + win_len]
            total += score_sequence(seq, win_len, ai_piece, opp)
            total -= score_sequence(seq, win_len, opp, ai_piece)
        return total

    for i in range(n):
        score += eval_line(board[i])
    for j in range(n):
        score += eval_line([board[i][j] for i in range(n)])
    for i in range(n):
        for j in range(n):
            if i + win_len <= n and j + win_len <= n:
                score += eval_line([board[i+k][j+k] for k in range(win_len)])
            if i + win_len <= n and j - win_len + 1 >= 0:
                score += eval_line([board[i+k][j-k] for k in range(win_len)])

    center = n // 2
    center_bonus = 3
    if board[center][center] == ai_piece:
        score += center_bonus
    elif board[center][center] == opp:
        score -= center_bonus

    return score


pruned_count = 0

def alpha_beta(board, depth, max_depth, alpha, beta, is_maximizing, ai_piece):

    global pruned_count

    w = check_winner(board)
    opp_piece = O if ai_piece == X else X
    if w == ai_piece:
        return 10000 - depth
    if w == opp_piece:
        return -10000 + depth
    if terminal(board):
        return 0
    if depth >= max_depth:
        return heuristic(board, ai_piece)

    if is_maximizing:
        max_eval = -math.inf
        for action in actions_ordered(board):
            eval_val = alpha_beta(result(board, action), depth + 1, max_depth,
                                  alpha, beta, False, ai_piece)
            max_eval = max(max_eval, eval_val)
            alpha = max(alpha, eval_val)
            if beta <= alpha:
                pruned_count += 1
                break
        return max_eval
    else:
        min_eval = math.inf
        for action in actions_ordered(board):
            eval_val = alpha_beta(result(board, action), depth + 1, max_depth,
                                  alpha, beta, True, ai_piece)
            min_eval = min(min_eval, eval_val)
            beta = min(beta, eval_val)
            if beta <= alpha:
                pruned_count += 1
                break
        return min_eval

def find_best_move(board, ai_piece):

    global pruned_count
    pruned_count = 0

    n = len(board)
    max_depth = get_max_depth(n)
    best_action = None
    best_val = -math.inf

    for action in actions_ordered(board):
        val = alpha_beta(result(board, action), 0, max_depth,
                         -math.inf, math.inf, False, ai_piece)
        if val > best_val:
            best_val = val
            best_action = action

    return best_action, best_val, pruned_count

def print_board(board):
    n = len(board)
    cw = 3 if n >= 10 else 2
    print("   " + " ".join(str(j).rjust(cw) for j in range(n)))
    print("   " + "-" * (cw * n + n - 1))
    for i, row in enumerate(board):
        cells = " ".join((c if c else ".").rjust(cw) for c in row)
        print(f"{i:2d} | {cells}")
    print()

def main():
    print("=" * 55)
    print("   TICTACTOE NxN - ALPHA-BETA PRUNING (DEPTH-LIMITED)")
    print("=" * 55)

    try:
        n = int(input("Nhập kích thước bảng N (vd: 3, 5, 10): "))
        if n < 3:
            n = 3
    except ValueError:
        n = 3

    win_len = get_win_length(n)
    max_depth = get_max_depth(n)
    print(f"\n>> Bảng: {n}x{n} | Cần {win_len} quân liên tiếp | Độ sâu: {max_depth}")
    print(f">> Move Ordering: Ưu tiên ô gần trung tâm (tối ưu hiệu quả cắt tỉa)\n")

    human = ""
    while human not in [X, O]:
        human = input("Bạn chọn X hay O? ").strip().upper()
    ai = O if human == X else X
    print(f">> Bạn: {human}  |  AI: {ai}  (X đi trước)\n")

    board = initial_state(n)
    total_pruned = 0

    while not terminal(board):
        print_board(board)
        current = player(board)

        if current == human:
            print(f"Lượt của bạn ({human}):")
            while True:
                try:
                    r = int(input(f"  Hàng (0-{n-1}): "))
                    c = int(input(f"  Cột (0-{n-1}): "))
                    if 0 <= r < n and 0 <= c < n and board[r][c] == EMPTY:
                        board = result(board, (r, c))
                        break
                    else:
                        print("  >> Ô không hợp lệ!")
                except ValueError:
                    print("  >> Vui lòng nhập số nguyên!")
        else:
            print(f"Lượt của AI ({ai}) - Alpha-Beta Pruning (sâu {max_depth})...")
            move, score, pruned = find_best_move(board, ai)
            total_pruned += pruned
            if move:
                board = result(board, move)
                print(f"  >> AI đánh: hàng {move[0]}, cột {move[1]}")
                print(f"  >> Điểm heuristic: {score} | Số lần cắt tỉa: {pruned}")
            else:
                break

    print_board(board)
    w = check_winner(board)
    print("=" * 55)
    if w == human:
        print("CHÚC MỪNG! Bạn thắng!")
    elif w == ai:
        print("AI thắng!")
    else:
        print("Hòa!")
    print(f"  Tổng số lần cắt tỉa: {total_pruned}")
    print("=" * 55)

if __name__ == "__main__":
    main()
```

Bài 5. Ứng dụng kĩ thuật cắt tỉa Alpha-beta vào game cờ tướng, cờ vua và cờ vây.


```
import math
import copy


RED   = 'red'
BLACK = 'black'
EMPTY = '.'

PIECES = {
    'K': (RED, 'Tướng'),  'k': (BLACK, 'Tướng'),
    'R': (RED, 'Xe'),     'r': (BLACK, 'Xe'),
    'N': (RED, 'Mã'),     'n': (BLACK, 'Mã'),
    'C': (RED, 'Pháo'),   'c': (BLACK, 'Pháo'),
    'P': (RED, 'Tốt'),    'p': (BLACK, 'Tốt'),
    'A': (RED, 'Sĩ'),     'a': (BLACK, 'Sĩ'),
    'E': (RED, 'Tượng'),  'e': (BLACK, 'Tượng'),
}


def initial_board():

    board = [
        ['r', 'n', 'e', 'a', 'k', 'a', 'e', 'n', 'r'],
        ['.', '.', '.', '.', '.', '.', '.', '.', '.'],
        ['.', 'c', '.', '.', '.', '.', '.', 'c', '.'],
        ['p', '.', 'p', '.', 'p', '.', 'p', '.', 'p'],
        ['.', '.', '.', '.', '.', '.', '.', '.', '.'],
        ['.', '.', '.', '.', '.', '.', '.', '.', '.'],
        ['P', '.', 'P', '.', 'P', '.', 'P', '.', 'P'],
        ['.', 'C', '.', '.', '.', '.', '.', 'C', '.'],
        ['.', '.', '.', '.', '.', '.', '.', '.', '.'],
        ['R', 'N', 'E', 'A', 'K', 'A', 'E', 'N', 'R'],
    ]
    return board

def get_color(piece):
    if piece == EMPTY or piece == '.': return None
    return PIECES.get(piece, (None, None))[0]

def get_opponent(color):
    return BLACK if color == RED else RED


def in_board(r, c):
    return 0 <= r <= 9 and 0 <= c <= 8

def in_palace(r, c, color):

    if color == RED:
        return 7 <= r <= 9 and 3 <= c <= 5
    else:
        return 0 <= r <= 2 and 3 <= c <= 5

def get_moves(board, r, c):

    piece = board[r][c]
    if piece == '.' or piece == EMPTY:
        return []
    color = get_color(piece)
    moves = []
    p = piece.upper()

    if p == 'K':   moves = _king_moves(board, r, c, color)
    elif p == 'A': moves = _advisor_moves(board, r, c, color)
    elif p == 'E': moves = _elephant_moves(board, r, c, color)
    elif p == 'R': moves = _rook_moves(board, r, c, color)
    elif p == 'N': moves = _knight_moves(board, r, c, color)
    elif p == 'C': moves = _cannon_moves(board, r, c, color)
    elif p == 'P': moves = _pawn_moves(board, r, c, color)

    return moves

def _king_moves(board, r, c, color):

    moves = []
    for dr, dc in [(-1,0),(1,0),(0,-1),(0,1)]:
        nr, nc = r+dr, c+dc
        if in_palace(nr, nc, color):
            target = board[nr][nc]
            if target == '.' or get_color(target) != color:
                moves.append((nr, nc))
    return moves

def _advisor_moves(board, r, c, color):

    moves = []
    for dr, dc in [(-1,-1),(-1,1),(1,-1),(1,1)]:
        nr, nc = r+dr, c+dc
        if in_palace(nr, nc, color):
            target = board[nr][nc]
            if target == '.' or get_color(target) != color:
                moves.append((nr, nc))
    return moves

def _elephant_moves(board, r, c, color):

    moves = []
    for dr, dc in [(-2,-2),(-2,2),(2,-2),(2,2)]:
        nr, nc = r+dr, c+dc
        if not in_board(nr, nc): continue

        if color == RED and nr < 5: continue
        if color == BLACK and nr > 4: continue

        mr, mc = r + dr//2, c + dc//2
        if board[mr][mc] != '.': continue
        target = board[nr][nc]
        if target == '.' or get_color(target) != color:
            moves.append((nr, nc))
    return moves

def _rook_moves(board, r, c, color):

    moves = []
    for dr, dc in [(-1,0),(1,0),(0,-1),(0,1)]:
        nr, nc = r+dr, c+dc
        while in_board(nr, nc):
            target = board[nr][nc]
            if target == '.':
                moves.append((nr, nc))
            elif get_color(target) != color:
                moves.append((nr, nc))
                break
            else:
                break
            nr += dr; nc += dc
    return moves

def _knight_moves(board, r, c, color):
    moves = []
    for dr, dc, br, bc in [(-2,1,-1,0),(-2,-1,-1,0),(2,1,1,0),(2,-1,1,0),
                            (-1,2,0,1),(-1,-2,0,-1),(1,2,0,1),(1,-2,0,-1)]:
        nr, nc = r+dr, c+dc
        block_r, block_c = r+br, c+bc
        if not in_board(nr, nc): continue
        if board[block_r][block_c] != '.': continue
        target = board[nr][nc]
        if target == '.' or get_color(target) != color:
            moves.append((nr, nc))
    return moves

def _cannon_moves(board, r, c, color):

    moves = []
    for dr, dc in [(-1,0),(1,0),(0,-1),(0,1)]:
        nr, nc = r+dr, c+dc
        jumped = False
        while in_board(nr, nc):
            target = board[nr][nc]
            if not jumped:
                if target == '.':
                    moves.append((nr, nc))
                else:
                    jumped = True
            else:
                if target != '.':
                    if get_color(target) != color:
                        moves.append((nr, nc))
                    break
            nr += dr; nc += dc
    return moves

def _pawn_moves(board, r, c, color):

    moves = []
    if color == RED:
        forward = -1
        crossed = r <= 4
    else:
        forward = 1
        crossed = r >= 5

    nr, nc = r + forward, c
    if in_board(nr, nc):
        target = board[nr][nc]
        if target == '.' or get_color(target) != color:
            moves.append((nr, nc))

    if crossed:
        for dc in [-1, 1]:
            nr, nc = r, c + dc
            if in_board(nr, nc):
                target = board[nr][nc]
                if target == '.' or get_color(target) != color:
                    moves.append((nr, nc))
    return moves


def find_king(board, color):
    target = 'K' if color == RED else 'k'
    for r in range(10):
        for c in range(9):
            if board[r][c] == target:
                return (r, c)
    return None

def is_terminal(board):

    red_king = find_king(board, RED)
    black_king = find_king(board, BLACK)
    return red_king is None or black_king is None

def get_winner(board):
    if find_king(board, RED) is None:   return BLACK
    if find_king(board, BLACK) is None: return RED
    return None

def all_moves(board, color):

    moves = []
    for r in range(10):
        for c in range(9):
            if get_color(board[r][c]) == color:
                for (nr, nc) in get_moves(board, r, c):
                    moves.append((r, c, nr, nc))
    return moves

def make_move(board, move):

    r, c, nr, nc = move
    new_board = copy.deepcopy(board)
    new_board[nr][nc] = new_board[r][c]
    new_board[r][c] = '.'
    return new_board


PIECE_VALUES = {
    'K': 10000, 'k': 10000,
    'R': 500,   'r': 500,
    'C': 300,   'c': 300,
    'N': 270,   'n': 270,
    'E': 120,   'e': 120,
    'A': 120,   'a': 120,
    'P': 30,    'p': 30,
    '.': 0,
}

def evaluate(board):

    score = 0
    for r in range(10):
        for c in range(9):
            p = board[r][c]
            if p != '.':
                val = PIECE_VALUES.get(p, 0)
                if get_color(p) == RED:
                    score += val
                else:
                    score -= val
    return score


MAX_DEPTH = 3

def alpha_beta(board, depth, alpha, beta, is_maximizing):
    """
    Alpha-Beta Pruning cho Cờ Tướng.
    MAX = ĐỎ, MIN = ĐEN.
    """
    winner = get_winner(board)
    if winner == RED:   return 100000 - depth
    if winner == BLACK: return -100000 + depth
    if depth == 0:      return evaluate(board)

    color = RED if is_maximizing else BLACK
    moves = all_moves(board, color)

    if not moves:
        return evaluate(board)

    def move_priority(mv):
        _, _, nr, nc = mv
        target = board[nr][nc]
        return -PIECE_VALUES.get(target, 0)
    moves.sort(key=move_priority)

    if is_maximizing:
        max_eval = -math.inf
        for move in moves:
            new_board = make_move(board, move)
            val = alpha_beta(new_board, depth - 1, alpha, beta, False)
            max_eval = max(max_eval, val)
            alpha = max(alpha, val)
            if beta <= alpha:
                break
        return max_eval
    else:
        min_eval = math.inf
        for move in moves:
            new_board = make_move(board, move)
            val = alpha_beta(new_board, depth - 1, alpha, beta, True)
            min_eval = min(min_eval, val)
            beta = min(beta, val)
            if beta <= alpha:
                break
        return min_eval

def find_best_move_chess(board, ai_color):
    """Tìm nước đi tốt nhất cho AI."""
    is_max = (ai_color == RED)
    best_move = None
    best_val = -math.inf if is_max else math.inf
    moves = all_moves(board, ai_color)

    if not moves:
        return None

    for move in moves:
        new_board = make_move(board, move)
        val = alpha_beta(new_board, MAX_DEPTH - 1, -math.inf, math.inf, not is_max)
        if is_max and val > best_val:
            best_val = val; best_move = move
        elif not is_max and val < best_val:
            best_val = val; best_move = move

    return best_move


PIECE_DISPLAY = {
    'K': '帥', 'k': '將',
    'R': '俥', 'r': '車',
    'N': '傌', 'n': '馬',
    'C': '炮', 'c': '砲',
    'P': '兵', 'p': '卒',
    'A': '仕', 'a': '士',
    'E': '相', 'e': '象',
    '.': '．',
}

def print_board(board):
    print()
    print("    " + "  ".join(str(c) for c in range(9)))
    print("   +" + "--" * 9 + "-+")
    for r in range(10):
        row_str = " | ".join(PIECE_DISPLAY.get(board[r][c], '？') for c in range(9))
        print(f" {r:2d}| {row_str} |")
        if r == 4:
            print("   |" + "~" * 19 + "|  ← Sông")
    print("   +" + "--" * 9 + "-+")
    print("    CHỮ HOA = ĐỎ (bạn/AI)   chữ thường = ĐEN")
    print()

def parse_move(input_str):

    try:
        parts = input_str.strip().split()
        r1, c1, r2, c2 = int(parts[0]), int(parts[1]), int(parts[2]), int(parts[3])
        return (r1, c1, r2, c2)
    except:
        return None

def main():
    print("=" * 60)
    print("      CỜ TƯỚNG - ALPHA-BETA PRUNING AI")
    print("=" * 60)
    print("Quân ĐỎ (CHỮ HOA): K=Tướng R=Xe N=Mã C=Pháo P=Tốt A=Sĩ E=Tượng")
    print("Quân ĐEN (chữ nhỏ): k=tướng r=xe n=mã c=pháo p=tốt a=sĩ e=tượng")
    print(f"AI độ sâu tìm kiếm: {MAX_DEPTH}\n")

    choice = ""
    while choice not in ['1', '2']:
        print("1. Bạn chơi ĐỎ (đi trước)")
        print("2. Bạn chơi ĐEN (đi sau)")
        choice = input("Chọn: ").strip()

    if choice == '1':
        human_color = RED
        ai_color = BLACK
    else:
        human_color = BLACK
        ai_color = RED

    print(f"\n>> Bạn: {'ĐỎ' if human_color == RED else 'ĐEN'}  |  AI: {'ĐỎ' if ai_color == RED else 'ĐEN'}")
    print(">> Nhập nước đi theo format: hàng1 cột1 hàng2 cột2 (vd: 9 0 8 0)\n")

    board = initial_board()
    current_color = RED

    while not is_terminal(board):
        print_board(board)
        color_name = 'ĐỎ' if current_color == RED else 'ĐEN'
        print(f"--- Lượt của: {color_name} ---")

        if current_color == human_color:
            while True:
                move_input = input("  Nhập nước đi (hàng1 cột1 hàng2 cột2): ").strip()
                move = parse_move(move_input)
                if move is None:
                    print("  >> Định dạng sai! Ví dụ: 9 0 8 0")
                    continue
                r1, c1, r2, c2 = move
                if not in_board(r1, c1) or not in_board(r2, c2):
                    print("  >> Tọa độ ngoài bàn cờ!")
                    continue
                if get_color(board[r1][c1]) != human_color:
                    print("  >> Đây không phải quân của bạn!")
                    continue
                valid = get_moves(board, r1, c1)
                if (r2, c2) not in valid:
                    print("  >> Nước đi không hợp lệ!")
                    continue
                board = make_move(board, move)
                break
        else:
            print(f"  AI ({color_name}) đang tính toán (Alpha-Beta sâu {MAX_DEPTH})...")
            move = find_best_move_chess(board, ai_color)
            if move:
                r1, c1, r2, c2 = move
                piece_name = PIECES.get(board[r1][c1], ('?', '?'))[1]
                print(f"  >> AI đi: {piece_name} ({r1},{c1}) → ({r2},{c2})")
                board = make_move(board, move)
            else:
                print("  >> AI không có nước đi! Thua!")
                break

        current_color = get_opponent(current_color)

    print_board(board)
    winner = get_winner(board)
    print("=" * 60)
    if winner == human_color:
        print("CHÚC MỪNG! Bạn thắng!")
    elif winner == ai_color:
        print("AI thắng!")
    else:
        print("  Trò chơi kết thúc!")
    print("=" * 60)

if __name__ == "__main__":
    main()
```
