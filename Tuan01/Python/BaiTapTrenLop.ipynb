13.	Viết hàm trộn 2 mảng một chiều thành 1 mảng một chiều với mỗi phần tử của mảng mới là tổng của 2 phần tương ứng từ 2 mảng cho trước. Trong quá trình trộn 2 mảng nếu mảng nào còn phần tử thì các phần tử còn lại của mảng đó sẽ đưa vào mảng mới.
```
Ví dụ:
Mảng a: 3 9 1 4
Mảng b: 2 7 4 3 2 8
Mảng kết quả: 5 16 5 7 2 8

```

````
def tron_mang_tong(a, b):
    kq = []
    i = j = 0
    n, m = len(a), len(b)

    while i < n and j < m:
        kq.append(a[i] + b[j])
        i += 1
        j += 1

    while i < n:
        kq.append(a[i])
        i += 1

    while j < m:
        kq.append(b[j])
        j += 1

    return kq

if __name__ == "__main__":
    a = list(map(int, input("Nhập mảng a: ").split()))
    b = list(map(int, input("Nhập mảng b: ").split()))
    print("Kết quả:", tron_mang_tong(a, b))
````

14.	Cho mảng 2 chiều a có m dòng, n cột chứa số nguyên, viết các hàm sau:
* a.	Tạo mảng a chứa các số nguyên ngẫu nhiên.
* b.	Xuất các phần tử thuộc dòng k.
* c.	Xuất các phần tử thuộc cột k.
* d.	Tìm dòng có tổng lớn nhất nhất.
* e.	Tìm cột có tích nhỏ nhất.
* f.	Xuất ra các phần tử thuộc dòng chẵn và cột lẻ trong a.
* g.	Tính trung bình cộng các phần tử chẵn thuộc dòng lẻ của a.
* h.	Tính trung bình cộng các phần tử thuộc biên.
* i.	Tính trung bình tích các phần tử không thuộc biên.

````
import random

def tao_mang_ngau_nhien(m, n, min_val=-10, max_val=10):
    mang = []
    for i in range(m):
        dong = [random.randint(min_val, max_val) for _ in range(n)]
        mang.append(dong)
    return mang

def xuat_mang(a):
    print("=== Mảng ngẫu nhiên ===")
    for dong in a:
        print(dong)
    print()

def xuat_dong_k(a, k):
    if 0 <= k < len(a):
        print(f"Dòng {k}: {a[k]}")
    else:
        print(f"Dòng {k} không tồn tại!")

def xuat_cot_k(a, k):
    if 0 <= k < len(a[0]):
        cot = [a[i][k] for i in range(len(a))]
        print(f"Cột {k}: {cot}")
    else:
        print(f"Cột {k} không tồn tại!")

def dong_tong_lon_nhat(a):
    dong_max = 0
    tong_max = sum(a[0])
    for i in range(1, len(a)):
        tong_hien_tai = sum(a[i])
        if tong_hien_tai > tong_max:
            tong_max = tong_hien_tai
            dong_max = i
    print(f"Dòng {dong_max} có tổng lớn nhất = {tong_max}")
    print(f"Nội dung dòng: {a[dong_max]}")

def cot_tich_nho_nhat(a):
    m, n = len(a), len(a[0])
    cot_min = 0
    tich_min = float('inf')

    for j in range(n):
        tich_cot = 1
        for i in range(m):
            tich_cot *= a[i][j]
        if tich_cot < tich_min:
            tich_min = tich_cot
            cot_min = j
    print(f"Cột {cot_min} có tích nhỏ nhất = {tich_min}")

def dong_chan_cot_le(a):
    print("Phần tử dòng chẵn, cột lẻ:")
    for i in range(0, len(a), 2):
        for j in range(1, len(a[0]), 2):
            print(f"  a[{i}][{j}] = {a[i][j]}")

def tb_chan_dong_le(a):
    ds_so_chan = []
    for i in range(1, len(a), 2):
        for gia_tri in a[i]:
            if gia_tri % 2 == 0:
                ds_so_chan.append(gia_tri)

    if not ds_so_chan:
        print("Trung bình phần tử chẵn ở dòng lẻ = 0 (Không có số chẵn)")
    else:
        tbc = sum(ds_so_chan) / len(ds_so_chan)
        print(f"Trung bình phần tử chẵn ở dòng lẻ = {tbc:.2f}  (gồm {len(ds_so_chan)} phần tử)")

def tb_bien(a):
    m, n = len(a), len(a[0])
    bien = []
    for i in range(m):
        for j in range(n):
            if i == 0 or i == m - 1 or j == 0 or j == n - 1:
                bien.append(a[i][j])
    tbc = sum(bien) / len(bien)
    print(f"Trung bình phần tử biên = {tbc:.2f}")

def tb_tich_khong_bien(a):
    m, n = len(a), len(a[0])
    if m < 3 or n < 3:
        print("Mảng quá nhỏ, không có phần tử không biên.")
        return

    tich = 1
    dem = 0
    for i in range(1, m - 1):
        for j in range(1, n - 1):
            tich *= a[i][j]
            dem += 1

    print(f"Trung bình tích phần tử không biên = {tich / dem:.2f}")

if __name__ == "__main__":

    m = int(input("Nhập số dòng m = "))
    n = int(input("Nhập số cột  n = "))

    a = tao_mang_ngau_nhien(m, n)
    xuat_mang(a)

    k_dong = int(input("Nhập k để xuất dòng k = "))
    xuat_dong_k(a, k_dong)

    k_cot = int(input("Nhập k để xuất cột k = "))
    xuat_cot_k(a, k_cot)

    print("-" * 30)
    dong_tong_lon_nhat(a)
    cot_tich_nho_nhat(a)
    print()

    dong_chan_cot_le(a)
    print()

    tb_chan_dong_le(a)
    tb_bien(a)
    tb_tich_khong_bien(a)
````

15.	Một sinh viên gồm các thông tin sau:
```
Mã sinh viên là chuỗi có 10 ký tự
Tên là chuỗi tối đa 20 ký tự
Năm sinh là một số nguyên
Điểm trung bình là một số thực a.
```
* a.	Xây dựng cấu trúc SINHVIEN mô tả một sinh viên
* b.	Cho một mảng có n sinh viên.
* c.	Viết hàm cho biết có bao sinh viên đủ điều kiện lên lớp, biết rằng sinh viên đủ điều kiện lên lớp khi điểm trung bình lớn hơn hoặc bằng 5.
* d.	Xuất các sinh viên đủ 20 tuổi.
* e.	Đếm số sinh viên học hệ đại học, biết rằng sinh viên hệ DH có mã sinh viên chứa 2 ký tự DH ở vị trí 2,3 trong chuỗi. VD: 02DH0001.
* f.	Cho biết trong mảng có bao nhiêu sinh viên có tên «Lan»
* g.	Cho biết trong mảng có bao nhiêu sinh viên có họ «Phan»


````
class SinhVien:
    def __init__(self, ma, ten, nam_sinh, diem_tb):
        self.ma = ma
        self.ten = ten
        self.nam_sinh = nam_sinh
        self.diem_tb = diem_tb

def nhap_ds_sv(n):
    ds = []
    for i in range(n):
        print(f"\n--- Nhập sinh viên {i+1} ---")
        ma = input("Mã SV: ")
        ten = input("Tên: ")
        nam_sinh = int(input("Năm sinh: "))
        diem = float(input("Điểm TB: "))
        ds.append(SinhVien(ma, ten, nam_sinh, diem))
    return ds

def dem_len_lop(ds):
    return len([sv for sv in ds if sv.diem_tb >= 5])

def sinh_vien_20_tuoi(ds, nam_hien_tai):
    return [sv for sv in ds if nam_hien_tai - sv.nam_sinh >= 20]

def dem_he_dh(ds):
    return len([sv for sv in ds if sv.ma[2:4].upper() == "DH"])

def dem_ten_lan(ds):
    return len([sv for sv in ds if sv.ten.split()[-1].lower() == "lan"])

def dem_ho_phan(ds):
    return len([sv for sv in ds if sv.ten.split()[0].lower() == "phan"])

n = int(input("Nhập số lượng sinh viên: "))
ds = nhap_ds_sv(n)

print("\nSố sinh viên lên lớp:", dem_len_lop(ds))

nam_hien_tai = int(input("\nNhập năm hiện tại: "))
print("\nDanh sách sinh viên đủ 20 tuổi:")
ds20 = sinh_vien_20_tuoi(ds, nam_hien_tai)

for sv in ds20:
    print(f" - Mã: {sv.ma}, Tên: {sv.ten}, Năm sinh: {sv.nam_sinh}, Điểm TB: {sv.diem_tb}")

print("\nSố sinh viên hệ DH:", dem_he_dh(ds))
print("Số sinh viên tên Lan:", dem_ten_lan(ds))
print("Số sinh viên họ Phan:", dem_ho_phan(ds))
````
